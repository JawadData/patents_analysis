```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import to_date, lit

# Créer une session Spark
spark = SparkSession.builder \
    .appName("Lecture de CSV avec Spark") \
    .getOrCreate()

df_wipo = spark.read.csv("/content/big_data_project.wipo_nv.csv", header=True)
df_es = spark.read.csv("/content/big_data_project.ES_net.csv", header=True)
df_cana = spark.read.csv("/content/big_data_project.canadien.csv", header=True)
df_fpo = spark.read.csv("/content/big_data_project.FPO.csv", header=True)
df_gp = spark.read.csv("/content/big_data_project.GP.csv", header=True)

df_wipo = df_wipo.withColumn("Application_Date", to_date("Application_Date", "yyyy-MM-dd'T'HH:mm:ss.SSS'Z'"))
df_es = df_es.withColumn("Date_publication", to_date("Date_publication", "yyyy-MM-dd'T'HH:mm:ss.SSS'Z'"))
df_cana = df_cana.withColumn("Filled Date", to_date("Filled Date", "yyyy-MM-dd'T'HH:mm:ss.SSS'Z'"))
df_fpo = df_fpo.withColumn("Filing Date", to_date("Filing Date", "MM/dd/yyyy"))
df_gp = df_gp.withColumn("Filing Date", to_date("Filing Date", "yyyy-MM-dd'T'HH:mm:ss.SSS'Z'"))

df_wipo_titles = df_wipo.select("Title", "Abstract", "Inventors", "Application_Date", "Office").withColumn("Source", lit("Wipo"))
df_es_titles = df_es.select("Title", "Patent_abstract", "Inventors", "Date_publication", "Applicants_country").withColumn("Source", lit("Es"))
df_cana_titles = df_cana.select("Title", "Abstract", "Inventors", "Filled Date", "Country").withColumn("Source", lit("Cana"))
df_fpo_titles = df_fpo.select("Title", "Summary", "Inventors", "Filing Date", "Country").withColumn("Source", lit("Fpo"))
df_gp_titles = df_gp.select("Title", "Abstract", "Inventors", "Filing Date", "Country Name").withColumn("Source", lit("Gp"))

dfs_merged = df_wipo_titles.union(df_es_titles).union(df_cana_titles).union(df_fpo_titles).union(df_gp_titles)

dfs_merged = dfs_merged.withColumnRenamed("Office", "country")
dfs_merged.show(3)

```

```python
df_selected = df_gp.select("Country Code", "Country Name")

# Afficher les premières lignes de la DataFrame sélectionnée pour vérification
df_selected.show()
```

```python
df_selected = df_selected.dropDuplicates()
```

```python
from pyspark.sql.functions import col

# Renommer les colonnes de df_selected
df_selected = df_selected.withColumnRenamed("Country Name", "Country_Name")
df_selected = df_selected.withColumnRenamed("Country Code", "Country_Code")
```

```python

from pyspark.sql.functions import length

df_selected = df_selected.filter(length(df_selected["Country_Code"]) <= 2)

df_selected.show()
```

```python
df_selected_merg = dfs_merged.select("country")
df_selected_merg.show(1)
```

```python
from pyspark.sql.functions import col, coalesce

# Join the DataFrames on the country code
df_merged = df_selected_merg.join(df_selected, df_selected_merg["country"] == df_selected["Country_Code"], "left_outer")

# Replace country codes with country names, keeping null values as they are
df_merged = df_merged.withColumn("country", coalesce(df_selected["Country_Name"], df_selected_merg["country"]))

# Drop the redundant columns
df_merged = df_merged.drop("Country_Code", "Country_Name")

# Show the updated DataFrame
df_merged.show()
```

```python
from pyspark.sql.functions import col, coalesce

# Renommer les colonnes dans df_selected
df_selected = df_selected.withColumnRenamed("Country Code", "country_code") \
    .withColumnRenamed("Country Name", "country_name")

# Jointure pour remplacer les codes de pays par les noms
dfs_merged_with_names = dfs_merged.join(df_selected, dfs_merged["country"] == df_selected["country_code"], "left_outer") \
    .withColumn("country_name", coalesce(df_selected["country_name"], dfs_merged["country"])) \
    .drop("country_code", "country") \
    .withColumnRenamed("country_name", "country")

# Afficher les premières lignes pour vérification
dfs_merged_with_names.show()
```

```python

from pyspark.sql.functions import col

# Sélectionner les lignes où la source est "Es"
df_es_rows = dfs_merged_with_names.filter(col("Source") == "Es")

# Afficher les lignes sélectionnées
df_es_rows.show()
```

```python
from pyspark.sql.functions import col, floor, year
import matplotlib.pyplot as plt

dfs_merged_with_names = dfs_merged_with_names.withColumn("Application_Year", year("Application_Date"))

dfs_merged_with_periods = dfs_merged_with_names.withColumn("Periode", floor(col("Application_Year") / 10) * 10)

brevets_par_periode = dfs_merged_with_periods.groupBy("Periode").count()

brevets_par_periode_cleaned = brevets_par_periode.dropna(subset=["Periode"])

brevets_par_periode_pd = brevets_par_periode_cleaned.toPandas()

plt.figure(figsize=(12, 6))
plt.bar(brevets_par_periode_pd["Periode"], brevets_par_periode_pd["count"])
plt.xlabel("Période (Années)")
plt.ylabel("Nombre de brevets")
plt.title("Nombre de brevets par période (Tranches de 10 ans)")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()




```

```python

from pyspark.sql import SparkSession
from pyspark.sql.functions import split, explode, desc, col
from pyspark.sql.types import StringType
from pyspark.sql.functions import udf
import string
import nltk
from nltk.corpus import stopwords

nltk.download("stopwords")
stop_words_en = set(stopwords.words("english"))
stop_words_fr = set(stopwords.words("french"))

def clean_word(word):
    return word.translate(str.maketrans('', '', string.punctuation))

clean_word_udf = udf(lambda word: clean_word(word), StringType())

df_cleaned = dfs_merged_with_names.withColumn("words", split(dfs_merged_with_names["Abstract"], " ")) \
    .select(explode("words").alias("word")) \
    .withColumn("word", clean_word_udf("word")) \
    .filter((col("word") != "") & ~col("word").isin(stop_words_en | stop_words_fr))

df_word_count = df_cleaned.groupBy("word").count().orderBy(desc("count"))

df_word_count.show()
```

```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Convertissez le DataFrame Spark en DataFrame pandas
df_word_count_pandas = df_word_count.toPandas()

# Utilisez set_index et générez le nuage de mots
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(df_word_count_pandas.set_index('word')['count'])

# Affichez le nuage de mots
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis('off')
plt.show()

```
